```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')

Reports = ["data/"+f for f in listdir("data") if isfile(join("data", f))]
```

# Plot with Province/State

```python
Province = "Hubei"


ProvinceData = pd.DataFrame()
plt.figure(figsize=(20,10))

for Report in Reports:
    DayReport = pd.read_csv(Report)
    DailyData = DayReport.loc[DayReport['Province/State'] == Province]
    ProvinceData = ProvinceData.append(DailyData)
    
Days = range(0,len(ProvinceData))
plt.plot(Days,ProvinceData["Deaths"], color = "red",label='Deaths', linewidth=6)
plt.plot(Days,ProvinceData["Recovered"], color = "Green", label='Recovered', linewidth=6)
plt.plot(Days,ProvinceData["Confirmed"]-ProvinceData["Recovered"], color = "Orange", label='Existing', linewidth=6)
#plt.plot(Days,CardiovascularDiseases_58,label='Cardiovascular Diseases',color = 'purple', linestyle = "--")
plt.legend(fontsize=25)
plt.title(f"Covid in {Province}",fontsize= 50)
```
